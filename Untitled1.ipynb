```python
def feature_eng(df):
    df_copy = df.copy()
    #df_copy['exp+emp'] = df_copy['experience_level'] + df_copy['employment_type'] # had a negative impact on the outcome MAE =38598
    #df_copy['job+exp'] = df_copy['job_title'] + df_copy['experience_level'] #  had a positive impact on the outcome MAE = 38422
    #df_copy['salary+comp'] = df_copy['salary_currency'] + df_copy['company_location'] # had a positive impact on the outcome MAE = 38261
    df_copy['job+salary'] = df_copy['job_title'] + df_copy['salary_currency'] # had a positive impact on the outcome MAE = 38062
    # salary+comp & job+salary MAE : 38183
    # job+salary & job+exp     MAE : 38440
    # job+salary & exp+emp     MAE : 38386
    # It seems that using just one strengthens our prediction
    df_copy['work_year'] = df_copy['work_year'].map({2020:0,  #R2-SCORE 0.3532
                                                    2021:1,
                                                    2022:2,
                                                    2023:3})
    df_copy['remote_ratio'] =  df_copy['remote_ratio'].map({0:1,
                                                           50:2,
                                                           100:3})
    return df_copy
```

```python
def cleaning(x):
    title_keywords = ['Lead', 'Developer', 'Engineer', 'Manager', 'Programmer', 'Architect',
                      'Researcher', 'Principal','Product','Constultant','Director of','Business','Scientist','Analyst']
    removed_keywords = []

    for keyword in title_keywords:
        if keyword in x:
            x = x.replace(keyword, '')
            removed_keywords.append(keyword)

    return removed_keywords, x.strip()

# Örnek kullanım
FE_1['title'], FE_1['job'] = zip(*FE_1['job_title'].map(cleaning))
FE_1.loc[FE_1['job'] == 'ML','job'] = 'Machine Learning'
FE_1.title = FE_1.title.apply(lambda x: ','.join(x))
FE_1.title = FE_1.title.apply(lambda x: x.replace(',',''))
```

```python
X = FE_1.drop('salary_in_usd',axis=1)
y = FE_1['salary_in_usd']
split_index = int(len(X) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
def objective_catboost(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 50, 500),
        'depth': trial.suggest_int('depth', 4, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.1),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10),
        'random_strength': trial.suggest_float('random_strength', 0.1, 1.0),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0.0, 1.0),
        'border_count': trial.suggest_int('border_count', 1, 255),
        'verbose': False,
        'loss_function': 'MAE',
    }

    model = CatBoostRegressor(**params)
    model.fit(X_train, y_train, eval_set=(X_test, y_test), verbose_eval=False, cat_features=X.select_dtypes('object').columns.tolist())
    y_pred = model.predict(X_test)
    accuracy = mean_absolute_error(y_test, y_pred) 
    return accuracy

study_catboost = optuna.create_study(direction='minimize')
study_catboost.optimize(objective_catboost, n_trials=100)
print("CatBoost Best Trial:")
print(study_catboost.best_trial.params)
print("CatBoost Best MAE:", study_catboost.best_trial.value)
```
