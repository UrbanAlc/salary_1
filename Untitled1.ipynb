```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
df = df.sort_values(by='work_year').reset_index(drop=True)
```

```python
le = LabelEncoder()
for i in df.select_dtypes('object').columns:
    df[i] = le.fit_transform(df[[i]])
```

```python
X = df.drop(['salary','salary_in_usd'],axis=1)
y = df['salary_in_usd']
```

```python
x_train,x_test,y_train,y_test  = X[:3004],X[3004:],y[:3004],y[3004:]
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(x_train,y_train)
xgb_pred = xgb.predict(x_test)
r2_score(y_test,xgb_pred),mean_absolute_error(y_test,xgb_pred),mean_squared_error(y_test,xgb_pred)
```
